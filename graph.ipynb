```python
import matplotlib.pyplot as plt
import numpy as np
from collections import namedtuple
```

```python
lines = []
with open('lines.txt') as f:
    for line in f:
        if line.strip():
            lines.append(list(map(int, line.split()))[1:])
```

```python
Window = namedtuple("Window", ['start', 'end'])
window = Window(90, 100)
print(window)
for line in lines:
    plt.plot(np.arange(window.start, window.end), line[window.start:window.end])
```
